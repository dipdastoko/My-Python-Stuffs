```python
def square(num):
    return num**2
```

```python
my_nums=[1,2,3,4,5]
```

```python
map(square, my_nums)
```

```python
for item in map(square, my_nums):
    print(item)
```

```python
list(map(square,my_nums))
```

```python
def splicer(mystring):
    if len(mystring)%2==0:
        return 'EVEN'
    else:
        return mystring[0]
```

```python
names=['Andy','Eve','Sally']
```

```python
list(map(splicer,names))
```
