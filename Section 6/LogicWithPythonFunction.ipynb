```python
def even_check(number):
    result = number % 2 ==0
    return result
```

```python
even_check(21)
```

```python
even_check(20)
```

```python
def check_even_list(num_list):
    even_numbers = []
    for number in num_list:
        if number%2==0:
            even_numbers.append(number)
        else:
            pass
    return even_numbers
```

```python
check_even_list([1,3,5,8,9])
```
