```python
def myfunc(a,b):
    return sum((a,b))*0.05
```

```python
myfunc(40,60)
```

```python
def myfunc(*args):
    return sum(args)*0.05
```

```python
myfunc(40,60,100)
```

```python
def myfunc(**kwargs):
    if 'fruit' in kwargs:
        print("My fruit is {}".format(kwargs['fruit']))
```

```python
myfunc(fruit='Apple')
```

```python
def myfunc(*args,**kwargs):
    print(args)
    print(kwargs)
    print("I want {} {}".format(args[0],kwargs['food']))
```

```python
myfunc(10,20,30,food='egg',fruit='apple')
```
