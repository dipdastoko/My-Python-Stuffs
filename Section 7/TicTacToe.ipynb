```python
from IPython.display import clear_output
import random
test_board = ["#", " "," "," "," "," "," "," "," "," "]
def display_board(board):
    clear_output()
    print("   |   |   ")
    print(f" {board[7]} | {board[8]} | {board[9]} ")
    print("   |   |   ")
    print("----------")
    print("   |   |   ")
    print(f" {board[4]} | {board[5]} | {board[6]} ")
    print("   |   |   ")
    print("----------")
    print("   |   |   ")
    print(f" {board[1]} | {board[2]} | {board[3]} ")
    print("   |   |   ")
```

```python
def player_input():
    player1 = 'Wrong Choice'
    player2 = ""
    while player1 not in ["X","O"]:
        player1 = input("Player1 please choose your marker: X or O -- ").upper()
        if player1 not in ["X","O"]:
            print("Please choose X or O")
        else:
            if player1=="X":
                player2="O"
            else:
                player2="X"
    print(f"Player1 marker: {player1}")
    print(f"Player2 marker: {player2}")
    return [player1,player2]
```

```python
def place_marker(board,marker,position):
    board[position]=marker
    display_board(board)
```

```python
def win_check(board,mark):
    if board[7]==mark and board[8]==mark and board[9]==mark:
        return True
    elif board[4]==mark and board[5]==mark and board[6]==mark:
        return True
    elif board[1]==mark and board[2]==mark and board[3]==mark:
        return True
    elif board[7]==mark and board[4]==mark and board[1]==mark:
        return True
    elif board[8]==mark and board[5]==mark and board[2]==mark:
        return True
    elif board[9]==mark and board[6]==mark and board[3]==mark:
        return True
    elif board[7]==mark and board[5]==mark and board[3]==mark:
        return True
    elif board[1]==mark and board[5]==mark and board[9]==mark:
        return True
    else:
        return False
```

```python
def choose_first():
    choice = random.randint(1,2)
    print(f"Player{choice} will choose first")
    return choice
```

```python
def space_check(board,position):
    if board[position] == " ":
        return True
    else:
        return False
```

```python
def full_board_check(board):
    if " " in board:
        return False
    return True
```

```python
def player_choice(board):
    space_available = False
    while space_available == False:
        playerChoice = input("Where you want to put mark on the board. Please choose among (1-9): ")
        if playerChoice.isdigit():
            if int(playerChoice) in range(1,10):
                if space_check(board,int(playerChoice)):
                    print("Space Available")
                    space_available = True
                else:
                    print("Space not available")
            else:
                print("Please choose a from the range (1-9)")
        else:
            print("Please choose a numeric number from 1 to 9")
    return int(playerChoice)
```

```python
def replay():
    playAgain = 'user choice'
    while playAgain.lower() not in ('y','n'):
        playAgain = input("Want to play again: Y/N")
        print(playAgain) 
    return True if playAgain.lower() == 'y' else False
```

```python
print('Welcome to Tic Tac Toe!')

while True:
    board = ["#", " "," "," "," "," "," "," "," "," "]
    display_board(board)
    
    p1,p2 = player_input()
    
    print("Let's Toss -----")
    player_turn=choose_first()

    game_on = True

    while game_on:
        # Player1's turn.
        if player_turn==1:
            print("Player1's Turn")
            if full_board_check(board):
                print("Draw!")
                break
            position = player_choice(board)
            place_marker(board,p1,position)
            if win_check(board,p1):
                print(f"Player1({p1}) Wins!")
                break
            player_turn=2

        # Player2's turn.
        if player_turn==2:
            print("Player2's Turn")
            if full_board_check(board):
                print("Draw!")
                break
            position = player_choice(board)
            place_marker(board,p2,position)
            if win_check(board,p2):
                print(f"Player1({p2}) Wins!")
                break
            player_turn=1

    if not replay():
        break
```
