```python
mystring = 'hello'
```

```python
mylist = []
for letter in mystring:
    mylist.append(letter)
print(mylist)
```

```python
mylist = [letter for letter in mystring]
```

```python
mylist
```

```python
mylist = []
for x in [2,4,6]:
    for y in [10,20,30]:
        mylist.append(x*y)
mylist
```

```python
mylist = [x*y if x*y%10==0 else 'N/A' for x in [2,4,6] for y in [1,10,100]]
mylist
```
