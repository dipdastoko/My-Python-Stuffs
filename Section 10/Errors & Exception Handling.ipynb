```python
try:
    f = open('testfile','r')
    f.write('write a test line')
except TypeError:
    print('There was a type error!')
except OSError:
    print('Hey there is a OS Error!')
finally:
    print("I always run")
    
```

```python
def ask_for_int():
    while True:
        try:
            result = int(input('please provide a number: '))
        except:
            print('Opps! thats not a number')
            continue
        else:
            print('yes, thank you!')
            break
        finally:
            print('End of try, except and finally')
            print('I will always run at the end')
```

```python
ask_for_int()
```
