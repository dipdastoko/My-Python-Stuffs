```python
def square(num):
    return num**2
```

```python
nums = [1,2,3,4,5]
```

```python
for item in map(square,nums):
    print(item)
```

```python
list(map(square,nums))
```

```python
def splicer(mystring):
    if len(mystring)%2==0:
        return "Even"
    else:
        return mystring[0]
```

```python
names=["Tomal","Tuku","Probal"]
```

```python
list(map(splicer,names))
```
