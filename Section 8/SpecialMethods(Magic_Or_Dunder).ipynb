```python
class Book():
    def __init__(self,title,author,pages):
        self.title = title
        self.author = author
        self.pages = pages
    def __str__(self):
        return f"{self.title} by {self.author}"
    def __len__(self):
        return self.pages
```

```python
b = Book('Python Rocks',"Jose",200)
```

```python
print(b)
```

```python
len(b)
```
