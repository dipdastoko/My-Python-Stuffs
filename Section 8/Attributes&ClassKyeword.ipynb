```python
class Dog():
    #Class Object Attribute
    species = 'Mammal'
    def __init__(self, breed, name):
        self.breed = breed
        self.name = name

    #Operations/Actions
    def bark(self,number):
        print("WOOF! My name is {} and my number is {}".format(self.name,number))
```

```python
my_dog = Dog('Lab','Frankie')
```

```python
my_dog.species
```

```python
my_dog.name
```

```python
my_dog.bark(2)
```

```python
class Circle():
    #Class Object Attribute
    pi = 3.14

    def __init__(self,radius=1):
        self.radius = radius
        self.area = Circle.pi*self.radius**2

    #Method
    def get_circumfrenece(self):
        return 2 * Circle.pi * self.radius
```

```python
my_circle = Circle(5)
```

```python
my_circle.get_circumfrenece()
```

```python
my_circle.area
```

```python
class Animal():
    def __init__(self):
        print("Animal Created")
    def who_am_i(self):
        print("I am an Animal")
    def eating(self):
        print("I am Eating")
```

```python
my_animal = Animal()
```

```python
my_animal.who_am_i()
```

```python
my_animal.eating()
```

```python
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Dog Created")
```

```python
mydog = Dog()
```

```python
mydog.eating()
```
